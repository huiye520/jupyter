```python
import tushare as ts
import pandas as pd
from pandas import DataFrame,Series
import numpy as np

pro = ts.pro_api('041ede9f0ab33a1281e1ce757b4b1ca0da35972e065ab1d08f9f078e')
df = pro.daily(ts_code='601888.SH',start_date = '20190101')
df.to_csv('./中免.csv')
```

```python
import pandas as pd
df = pd.read_csv('./中免.csv')
#df.head()
df.drop(labels='Unnamed: 0',axis=1,inplace=True)
df.head()
df['trade_date'].dtype
df['time'] = pd.to_datetime(df['trade_date'],format = '%Y%m%d')
df.info()
df.head()
df.set_index('time',inplace=True)
df.head()
```

```python
# (收盘-开盘) / 开盘 >0.03

df= df.loc[(df['close'] - df['pre_close']) /df['pre_close'] > 0.03]
df.index
```

```python
#(df['open'] - df['close'].shift(1)) /df['close'].shift(1) < -0.02
df= df.loc[(df['open'] - df['close'].shift(1)) /df['close'].shift(1) < -0.02]
df.head()
```

```python
df.index
new_df = df.loc['2020-01':'2023-01']
new_df.head(2)
df_monthly = new_df.resample('M').first()
df_monthly
```

```python
cost = df_monthly['open'].sum()*100
cost
df_yearly = new_df.resample('A').last()[:-1]
resv = df_yearly['open'].sum()*1200
last_money = 400*new_df['close'][-1]
resv+ last_money - cost
```
